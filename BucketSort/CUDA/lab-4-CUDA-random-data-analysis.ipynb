```python
tk = th.Thicket.from_caliperreader(glob("Semester-Project/CUDA/test/Cali-Random/*.cali"))

tk.dataframe = tk.dataframe.drop(["nid", "spot.channel", "Min time/rank", "Max time/rank", "Avg time/rank"], axis = 1)
```

```python
tk.statsframe.dataframe["time"] = 1
print(tk.tree())
```

```python
gb = tk.groupby("InputSize")
```

```python
ctk = th.Thicket.concat_thickets(
    thickets=list(gb.values()),
    headers=list(gb.keys()),
    axis="columns",
    metadata_key="num_threads"
)
```

```python
ctk.dataframe
```

```python
# Drop unwanted columns
ctk.dataframe = ctk.dataframe.reset_index().drop([("node", "")], axis = 1)
# Rename index to remove tuple
ctk.dataframe = ctk.dataframe.rename({("name", ""): "name", ("num_threads", ""): "num_threads"}).set_index(["name", "num_threads"])
```

```python
ctk.dataframe
```

```python
function_names = set(ctk.dataframe.index.get_level_values("name"))
threads = sorted(list(set(ctk.dataframe.index.get_level_values("num_threads"))))
num_vals = sorted(list(set(ctk.dataframe.columns.get_level_values(0))))
```

```python
def plot_mpl(df1, df2, df3, df4, title, xlabel, ylabel):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.plot(
        df1,
        marker='o',
        color='tab:blue'
    )
    ax.plot(
        df2,
        marker='o',
        color='tab:orange'
    )
    ax.plot(
        df3,
        marker='o',
        color ='tab:green'
    )
    ax.plot(
        df4,
        marker='o',
        color ='tab:red'
    )
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.legend(["2^16 Values", 
               "2^18 Values",
               "2^20 Values",
               "2^22 Values"])
    plt.grid(True)
    # plt.yscale("log", base=10)
    plt.xscale("log", base=2)
    plt.xticks(threads)
    plt.show()
    return fig, ax

def plot_mpl2(df1, title, xlabel, ylabel):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.plot(
        df1,
        marker='o',
        color='tab:blue'
    )
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.legend(["Bucket sort"])
    plt.grid(True)
    # plt.yscale("log", base=10)
    plt.xscale("log", base=2)
    plt.xticks(threads)
    plt.show()
    return fig, ax
```

```python
names = []
vals = []

for name in function_names:
    names.append(name)
    for val in num_vals:
        vals.append(val)

print(names)
print(vals)
```
