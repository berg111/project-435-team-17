```python
def plot_mpl(df, title, xlabel, ylabel):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.plot(
        df,
        marker='o',
    )
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.legend(df.columns)
    plt.grid(True)
    # plt.yscale("log", base=10)
    plt.xscale("log", base=2)
    plt.xticks(threads)
    plt.show()
    return fig, ax
```

```python
array_sizes = []
for pair in array_sizes1:
    array_sizes.append(pair[0])
    
array_sizes = list(set(array_sizes))

print(array_sizes)
print(threads)
```

```python
for name in function_names:
    for asize in array_sizes:
        tcols = [col for col in ctk.dataframe.columns if col[0][0] == asize]
        plot_mpl(
            df = ctk.dataframe.loc[name][tcols],
            title = f"{name} size: {asize}",
            xlabel = "processes",
            ylabel = "Time (seconds)"
        )
```

```python
def plot_mpl1(df1, df2, df3, df4, df5, df6, df7, title, xlabel, ylabel):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.plot(
        df1,
        marker='o',
        color='tab:blue'
    )
    ax.plot(
        df2,
        marker='o',
        color='tab:orange'
    )
    ax.plot(
        df3,
        marker='o',
        color ='tab:green'
    )
    ax.plot(
        df4,
        marker='o',
        color ='tab:red'
    )
    ax.plot(
        df5,
        marker='o',
        color ='tab:purple'
    )
    ax.plot(
        df6,
        marker='o',
        color ='tab:grey'
    )
    ax.plot(
        df7,
        marker='o',
        color ='tab:pink'
    )
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.legend(["2^16", "2^18", "2^20", "2^22"])
    # ax.legend(df1.columns, df2.columns, df3.columns, df4.columns, df5.columns, df6.columns, df7.columns,)
    plt.grid(True)
    # plt.yscale("log", base=10)
    plt.xscale("log", base=2)
    plt.xticks(threads)
    plt.show()
    return fig, ax

def plot_mpl2(df1, df2, df3, df4, title, xlabel, ylabel):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.plot(
        df1,
        marker='o',
        color='tab:blue'
    )
    ax.plot(
        df2,
        marker='o',
        color='tab:orange'
    )
    ax.plot(
        df3,
        marker='o',
        color ='tab:green'
    )
    ax.plot(
        df4,
        marker='o',
        color ='tab:red'
    )
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.legend(["2^16", "2^18", "2^20", "2^22"])
    # ax.legend(df1.columns, df2.columns, df3.columns, df4.columns, df5.columns, df6.columns, df7.columns,)
    plt.grid(True)
    # plt.yscale("log", base=10)
    plt.xscale("log", base=2)
    plt.xticks(threads)
    plt.show()
    return fig, ax
```
